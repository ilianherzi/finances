```python
import pandas as pd
from enum import Enum
```

```python
pd.options.display.max_rows = 1000
```

```python
# pylint: disable-next:
class Accounts(Enum):
    CHASE_CREDIT_CARD: str = "CCC"
    WELLS_FARGO_CHECKING: str = "WFCh"
    WELLS_FARGO_CREDIT_CARD: str = "WFCC"
    FIRST_REPUBLIC_CHECKING: str = "FRC"
    FIRST_REPUBLIC_MORTGAGE: str = "FRM"
        
MONTHS = [
    pd.to_datetime(f"2023-{n}-01 12:00")
    for n in range(4, 11)
]
```

```python
df = pd.read_csv("./aggregated_statement.csv").iloc[:, 1:]
df["Date"] = pd.to_datetime(df["Date"], format="%m/%d/%Y")
mask = ((df["Date"] >= "2023-04-01 12:00"))
df = df[mask].sort_values("Date").reset_index()
```

```python
months_df = []
for month_start, month_end in zip(MONTHS[:-1], MONTHS[1:]):
    accounts = []
    for account in Accounts:
        mask = ((df["Date"] >= month_start) & (df["Date"] < month_end) & (df["Account"] == account.name))
        accounts.append(df[mask].reset_index(drop=True).iloc[:, 1:])
    months_df.append(accounts)
```
